Filter de multiobjecten uit de dataset en maak er een aparte csv van

```python
data = pd.read_csv("/content/naturalis_petrology_sample.csv", sep=';', header=None)
```

```python
data = data.iloc[1: , :]
```

```python
mediaData = data[[0, 36]]
```

```python
test = pd.DataFrame()

for index, row in mediaData.iterrows():
  # row['ignore'] = False

  if row[36] != row[36]:
    mediaData = mediaData.drop(index)
  else :
    if '|' in row[36]:
      row[36] = row[36].replace('|', ' ')

      array = row[36].split()
      mediaData = mediaData.drop(index)
      
      for mediaObject in array:
        row = {0: row[0], 36: mediaObject}

        test = test.append(row, ignore_index=True)

mediaData = mediaData.append(test, ignore_index=True)
```

```python
mediaData = mediaData.sort_values(by=[36])
```

```python
pd.set_option('display.max_rows', mediaData.shape[0]+1)
print(mediaData)
```

Sorteer op langste columns

```python
data = pd.read_csv("/content/utrecht_occurrence.csv", sep=';', header=None)
```

```python
columns = {}

for column in data:
  # columns[data[column][0]] = 0
  columns[column] = 0

for index, row in data.iterrows():
  # print(row)
  rowCounter = 0

  for subrow in row:
    # print(subrow)
    subrow = str(subrow)

    if columns[rowCounter] < len(subrow):
      columns[rowCounter] = len(subrow)

    rowCounter = rowCounter + 1

columnLengths = columns
```

```python
print(columnLengths)
```

Create query

```python
query = 'ALTER TABLE `TABLE 8`'

for column in data:
  query = query + ', MODIFY ' + data[column][0] + ' BLOB(' + str(columnLengths[column]) + ')'

# print(query)

createQuery = 'CREATE TABLE utrechtFossilOccurrences ( '

for column in data:
  createQuery = createQuery + '`' + data[column][0] + '`' + ' VARCHAR(' + str(columnLengths[column]) + '), '

createQuery = createQuery + ');'

print(createQuery)
```
